### Merging departments.csv with ords_prods_merge

```python
# Transpose departments_data
df_departments = df_departments.T

# Add an index column
df_departments.reset_index()

# Create a new header with the values of the first row
new_header = df_departments.iloc[0]

# Remove the first row
df_departments = df_departments[1:]

# Reset the header row with the values in 'new_header'
df_departments.columns = new_header
```

```python
# Convert 'department_id' in ords_prods_merge to a string
ords_prods_merge['department_id'] = ords_prods_merge['department_id'].astype('str')
```

```python
# Merge ords_prods_merge with df_departments
ords_prods_merge_dep = ords_prods_merge.merge(df_departments, left_on = 'department_id', right_index = True, indicator = False)
```

### Create a column that defines the 'order_day_of_week' column

```python
# Create a series of the days of the week
dow = {'dow_code' : [0, 1, 2, 3, 4, 5, 6], 'day' : ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']}
```

```python
# Convert that 'dow' series into a dataframe
df_dow = pd.DataFrame(dow)
```

```python
# Merge ords_prods_merge_dep_day with df_dow
ords_prods_merge_dep_day = ords_prods_merge_dep.merge(df_dow, left_on = 'order_day_of_week', right_on = 'dow_code', indicator = False)
```

```python
# Check the output after the two merges
ords_prods_merge_dep_day.head(5)
```

## 04. Data Wrangling & Consistency Checks

### Rename columns to be uniform

```python
# Create a new dataframe with the new column names
df_customers_uniform = df_customers.rename({'First Name':'first_name', 'Surnam':'last_name', 'Gender':'gender', 'STATE':'state', 'Age':'age', 'n_dependants':'number_of_dependents', 'fam_status':'family_status'}, axis = 1)
```

```python
# Check the output
df_customers_uniform.head(15)
```

### Check the data types of the columns in the dataframe

```python
# Describe the columns with integer values
df_customers_uniform.describe()
```

```python
# Check the data types of the columns
df_customers_uniform.dtypes
```

```python
# Convert the user_id column from an integer to a string
df_customers_uniform['user_id'] = df_customers_uniform['user_id'].astype('str')
```

```python
# Check the data type of the user_id column
df_customers_uniform['user_id'].dtype
```

### Check for mixed-typed data

```python
# Check for mixed types
for col in df_customers_uniform.columns.tolist():
  weird = (df_customers_uniform[[col]].map(type) != df_customers_uniform[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers_uniform[weird]) > 0:
    print (col)
```

```python
# Convert the first_name column from a mixed-type to a string
df_customers_uniform['first_name'] = df_customers_uniform['first_name'].astype('str')
```

### Check for missing values

```python
# Check for missing values
df_customers_uniform.isnull().sum()
```

### Check for duplicated rows

```python
# Create a subset of the dataframe with all duplicated rows
df_dups = df_customers_uniform[df_customers_uniform.duplicated()]
```

```python
# Check to see if there are duplicates
df_dups
```

## 05. Merging Dataframes

### Verify that the key columns are the same data type

```python
# Checking the data type of the user_id column in our previously prepared data
ords_prods_merge_dep_day['user_id'].dtype
```

```python
# Converting the user_id column of our previously prepared data to a string
ords_prods_merge_dep_day['user_id'] = ords_prods_merge_dep_day['user_id'].astype('str')
```

### Merge the two dataframes

```python
# Use an inner merge to merge the customer data with the previously prepared data, using the user_id as a key column
complete_instacart_data = ords_prods_merge_dep_day.merge(df_customers_wrangled, on = 'user_id', how = 'inner', indicator = True)
```

```python
# Check the output
complete_instacart_data.head(5)
```

```python
# Check the merge flag to verify if it was a full match
complete_instacart_data['_merge'].value_counts(dropna = False)
```

### Verify that the merge was a full match

```python
# Use an outer merge to merge the two dataframes
complete_instacart_data_check = ords_prods_merge_dep_day.merge(df_customers_wrangled, on = 'user_id', how = 'outer', indicator = True)
```

```python
# Check the merge flag to verify if it was a full match
complete_instacart_data_check['_merge'].value_counts(dropna = False)
```

### Drop the merge flag from the dataframe

```python
# Drop the merge flag column
complete_instacart_data = complete_instacart_data.drop(columns = ['_merge'])
```
